```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data1.csv file
x = []
with open('data1.csv', 'r') as data1:
    reader = csv.reader(data1, delimiter=',')
    for row in reader:
        if row[0] == 'x1': # remove first line of the file
            continue
        else:
            x.append([row[0], row[1], -1, 0]) # [x, y, probability, distribution_label=0]
        
X = np.array(x, dtype=np.float128)
```

```python
# plot raw data from data1.csv
plt.plot(X[:,0], X[:, 1], 'ko')
plt.title('Raw Data')
plt.show()
```

```python
def p(x, d, mu, sigma):
    """
    The probability of x in normal distribution (mu, sigma)
    
    x: the input vector (1 x d)
    d: the input size
    mu: the mu of normal distribution
    sigma: the covariance of normal distribution
    Return: the probability of x in normal-distribution(mu, sigma)
    """
    return (1/(np.sqrt(((2*np.pi)**d)*(np.linalg.det(sigma)))))*np.exp((-1/2)*np.transpose(x-mu)@np.linalg.inv(sigma)@(x-mu))
    
def classify(x):
    """
    Assign label to each data in the dataset meaning whether they belong to distribution 1 or 2
    
    x: the data set matrix (m x 3) [[x1, x2, probability, distribution_label], ...]
    Return: update the distribution_label and probability of input matrix
    """
    mu1 = [5, 5]
    mu2 = [2, 2]
    sigma = [[1, 0], [0, 1]]
    
    probability1 = p(x[:2], 2, mu1, sigma) # x[:2] = [x1, x2]
    probability2 = p(x[:2], 2, mu2, sigma)
    
    if probability1 > probability2:
        x[3] = 1
    else:
        x[3] = 2
    x[2] = max(probability1, probability2)
```

```python
# for each data , find its class (whether belong to distribution 1 or 2)
np.apply_along_axis(classify, 1, X)

# separate each distribution data
dist1 = []
dist2 = []
for i in X:
    if i[3] == 1:
        dist1.append(i[:3])
    else: 
        dist2.append(i[:3])

        
dist1 = np.array(dist1, dtype=np.float128)
dist2 = np.array(dist2, dtype=np.float128)
```

```python
def L(x):
    """
    The Likelihood function
    
    x: the data set matrix (m x 3)
    Return: the likelihood of x
    """
    return np.prod(x[:, 2]) # x[:, 2] = probabilities
    
def LL(x):
    """
    The Log-Likelihood function
    
    x: the data set matrix (m x 3)
    Return: the Log-Likelihood of x
    """
    return np.sum(np.log(x[:, 2])) # x[:, 2] = probabilities
```

ุจุฑุง ฺฉ ุฏุชุงุณุช $ X_{m \times n} $ ฺฉู ุดุงูู ฺูุฏู ุฏุงุฏู ุงุณุช, ุฎุฑูุฌ ฺฉ ูฺฉุชูุฑ ุงุณุช $ h_{m \times 1} $ ฺฉู ุญุงู ููุฏุงุฑ ุชุฎูู ุฒุฏู ุดุฏู ุจุฑุง ูุฑ ุฏุงุฏู ุงุณุช:
$$ \small{h_{w,b}(X) = Xw + b} $$
    <h5>

    </h5>
ุชุงุจุน ูุฒูู ุจุฑุง ูุฏู ุฑฺฏุฑุณูู ุฎุท ุชุงุจุน MSE ุงุณุช ฺฉู ุจุฑุง ฺฉ ูุฌููุนู ุงุฒ ุฏุงุฏู ูุง ุฏุงุฑู:

    </h5>
ุงูฺฏูุฑุชู ูุฒูู ฺฏุฑุงุฏุงู ุ ุฑูุด ุจุฑุง ูููู ฺฉุฑุฏู ฺฉ ุชุงุจุน ุงุณุช ุจุง ุดุฑูุน ุงุฒ ฺฉ ุฌุง ุงุฒ ุชุงุจุน ู ุจุฑุฏุงุดุชู ฺฏุงู ูุง ฺฉูฺฺฉ ุจุฑุงุณุงุณ ฺฏุฑุงุฏุงู ุชุงุจุน ุชุง ุฑุณุฏู ุจู ููุทู ุจููู.
ุฏุฑ ุฑฺฏุฑุณูู ุฎุท, ุชุงุจุน ฺฉู ูุง ุชูุงุด ูฺฉูู ฺฉููู ฺฉูู ููุงู ุชุงุจุน ูุฒูู $ J_{w,b}(X) $ ุงุณุช. ูุดุชูุงุช ุฌุฒุฆ ุชุงุจุน ูุฒูู ุนุจุงุฑุช ุงูุฏ ุงุฒ:

$$ 
\small{
\begin{equation}
   \begin{split}
       \frac{\partial J_{w,b}(X)}{\partial w_j} &= \frac{\partial}{\partial w_j} \frac{1}{2m}\sum_{i=1}^m (h_{w,b}(x^{(i)}) - y^{(i)})^2 \\
       &= \frac{1}{2m}\sum_{i=1}^m 2 (h_{w,b}(x^{(i)}) - y^{(i)}) \frac{\partial}{\partial w_j} h_{w,b}(x^{(i)}) \\
       &= \frac{1}{m}\sum_{i=1}^m (h_{w,b}(x^{(i)}) - y^{(i)}) x_{j}^{(i)} \\
       &= \frac{1}{m}\sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)}) x_{j}^{(i)}
   \end{split}
   \end{equation}
}   
$$
   
$$
\small{
\begin{equation}
   \begin{split}
       \frac{\partial J_{w,b}(X)}{\partial b} &= \frac{1}{m}\sum_{i=1}^m 
       (h_{w,b}(x^{(i)}) - y^{(i)}) \frac{\partial}{\partial b} h_{w,b}(x^{(i)}) \\
       &= \frac{1}{m}\sum_{i=1}^m (h_{w,b}(x^{(i)}) - y^{(i)}) \\
       &= \frac{1}{m}\sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)})
   \end{split}
   \end{equation}
}
$$
   
- ุฏุฑุญููุช ุงู ุฏู ูุดุชู ฺฉุณุงู ูุณุชูุฏ ุจู ุงู ุฏูู ฺฉู ุฏุฑ ุฏูู $ x_{0}^{(i)} = 1 $.

```python
class GD:
    def h(self, X, w, b):
        """
        The hypothesis function

        X: the data set matrix (m x n)
        w: the weights vector (n x 1)
        b: the bias (1 x 1)

        Return: a vector stacking predictions for all data items (m x 1)
        """
        return X@w + b

    def MSE(self, y_true, y_pred):
        """
        Mean squared error cost function

        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)

        Return: a single number representing the cost
        """
        return (1/(2*len(y_true)))*np.sum((y_pred-y_true)**2)
    
    def gradient(self, X, y_true, y_pred):
        """
        The gradient of cost function

        X: the data set matrix (m x n)
        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)

        Return: vector dJ/dw (n x 1) and number dJ/db (1 x 1)
        """
        m = len(y_true)
        dJ_dw = (1/m)*(X.T@(y_pred - y_true))
        dJ_db = (1/m)*(np.sum(y_pred - y_true))
        return (dJ_dw, dJ_db)
    
    def update_parameters(self, X, y_true, y_pred, w, b, alpha):
        """
        This function updates parameters w and b according to their derivations.
        It should compute the cost function derivations with respect to w and b first,
            then take a step for each parameters in w and b.

        X: the data set matrix (m x n)
        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)
        w: the weights vector (n x 1)
        b: the bias (1 x 1)
        alpha: the learning rate

        Return: the updated parameters w and b
        """
        [dJ_dw, dJ_db] = self.gradient(X , y_true , y_pred) 
        w = w - alpha * dJ_dw 
        b = b - alpha * dJ_db
        return (w ,b)
    
    def gradient_descent(self, X, y, alpha = 0.0001, epoch = 50):
        """
        The gradient descent algorithm:
            1. initialize parameters w and b
            for i in epoch:
            2. compute the hypothesis h(X, w, b)
            3. update the parameters using function update_parameters(X, y_true, y_pred, w, b, alpha)
            4. compute the cost and see the cost is decreasing in each step (optional)

        X: the data set matrix (m x n)
        y: the vector of true labels of data items (m x 1)
        alpha: the learning rate
        epoch: number of steps gradient descent should take to converge

        Return: the best parameters w and b gradient descent found at last
        """
        speedLog = time()
        ww = [-0.15, 0.6]
        w = np.array(ww)
        b = -59.5
        costPerEpoch = np.zeros((epoch))
        for i in range(epoch):
            hyp = self.h(X, w, b)
            w, b = self.update_parameters(X, y, hyp, w, b, alpha)
            costPerEpoch[i] = self.MSE(y, hyp)
        speedLog = time() - speedLog
        return w, b, costPerEpoch, speedLog
```

```python
class SGD:
    def h(self, X, w, b):
        """
        The hypothesis function

        X: the data set matrix (m x n)
        w: the weights vector (n x 1)
        b: the bias (1 x 1)

        Return: a vector stacking predictions for all data items (m x 1)
        """
        return X@w + b

    def MSE(self, y_true, y_pred):
        """
        Mean squared error cost function

        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)

        Return: a single number representing the cost
        """
        print(y_pred)
        return (1/(2*len(y_true)))*np.sum((y_pred-y_true)**2)

    def stochastic_gradient_descent(self, X, y, alpha = 0.1, epoch = 4, batch_size = 1):
        """
        The Stochastic gradient descent algorithm:
            1. initialize parameters w and b
            for i in epoch:
            2. compute the hypothesis h(X, w, b)
            3. update the parameters
            4. compute the cost (optional)

        X: the data set matrix (m x n)
        y: the vector of true labels of data items (m x 1)
        alpha: the learning rate
        epoch: number of steps SGD should take to converge
        batch_size: the number of examples you look at before making a weight update. 

        Return: the best parameters w and b gradient descent found at last
        """    
        speedLog = time()
        ww = [-0.15, 0.6]
        w = np.array(ww)
        b = -59.5
        costPerEpoch = np.zeros((epoch))
        for i in range(epoch):

            indexes = np.random.randint(0, len(X), batch_size) # random sample

            Xs = X[indexes]
            ys = y[indexes]
            N = len(Xs)
            
            hyp = self.h(Xs, w, b) - ys

            # Updating parameters w and b    
            print(Xs)
            w = w - alpha * (1/N)*(Xs.T@(hyp - ys))
            b = b - alpha * (1/N)*(np.sum(hyp - ys))
            print(f'w_{i} = {w}')
            print(f'b_{i} = {b}')
            

            costPerEpoch[i] = self.MSE(ys, hyp)

        speedLog = time() - speedLog
        return w, b, costPerEpoch, speedLog
```

ุงูฺฏูุฑุชู ูุฒูู ฺฏุฑุงุฏุงู ูุณุฎู mini-batch ุ ุฑูุด ุจุฑุง ูููู ฺฉุฑุฏู ฺฉ ุชุงุจุน ุงุณุช ุจุง ุดุฑูุน ุงุฒ ฺฉ ุฌุง ุงุฒ ุชุงุจุน ู ุจุฑุฏุงุดุชู ฺฏุงู ูุง ฺฉูฺฺฉ ุจุฑุงุณุงุณ ฺฏุฑุงุฏุงู ุชุงุจุน ุชุง ุฑุณุฏู ุจู ููุทู ุจููู.
ุฏุฑ ุฑฺฏุฑุณูู ุฎุท, ุชุงุจุน ฺฉู ูุง ุชูุงุด ูฺฉูู ฺฉููู ฺฉูู ููุงู ุชุงุจุน ูุฒูู $ J_{w,b}(X) $ ุงุณุช. ูุดุชูุงุช ุฌุฒุฆ ุชุงุจุน ูุฒูู ุนุจุงุฑุช ุงูุฏ ุงุฒ:

$$ 
\small{
\begin{equation}
   \begin{split}
       \frac{\partial J_{w,b}(X)}{\partial w_j} &= \frac{\partial}{\partial w_j} \frac{1}{2N}\sum_{i=1}^N (h_{w,b}(x^{(i)}) - y^{(i)})^2 \\
       &= \frac{1}{2N}\sum_{i=1}^N 2 (h_{w,b}(x^{(i)}) - y^{(i)}) \frac{\partial}{\partial w_j} h_{w,b}(x^{(i)}) \\
       &= \frac{1}{N}\sum_{i=1}^N (h_{w,b}(x^{(i)}) - y^{(i)}) x_{j}^{(i)} \\
       &= \frac{1}{N}\sum_{i=1}^N (\hat{y}^{(i)} - y^{(i)}) x_{j}^{(i)}
   \end{split}
   \end{equation}
}   
$$
   
$$
\small{
\begin{equation}
   \begin{split}
       \frac{\partial J_{w,b}(X)}{\partial b} &= \frac{1}{N}\sum_{i=1}^N 
       (h_{w,b}(x^{(i)}) - y^{(i)}) \frac{\partial}{\partial b} h_{w,b}(x^{(i)}) \\
       &= \frac{1}{N}\sum_{i=1}^N (h_{w,b}(x^{(i)}) - y^{(i)}) \\
       &= \frac{1}{N}\sum_{i=1}^N (\hat{y}^{(i)} - y^{(i)})
   \end{split}
   \end{equation}
}
$$
   
- ุฏุฑุญููุช ุงู ุฏู ูุดุชู ฺฉุณุงู ูุณุชูุฏ ุจู ุงู ุฏูู ฺฉู ุฏุฑ ุฏูู $ x_{0}^{(i)} = 1 $.
- N ููุงู ุงูุฏุงุฒู Batch ูุง ุงุณุช.

```python
class MiniBatchGD:
    def h(self, X, w, b):
        """
        The hypothesis function

        X: the data set matrix (m x n)
        w: the weights vector (n x 1)
        b: the bias (1 x 1)

        Return: a vector stacking predictions for all data items (m x 1)
        """
        return X@w + b

    def MSE(self, y_true, y_pred):
        """
        Mean squared error cost function

        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)

        Return: a single number representing the cost
        """
        return (1/(2*len(y_true)))*np.sum((y_pred-y_true)**2)
    
    def gradient(self, X, y_true, y_pred):
        """
        The gradient of cost function

        X: the data set matrix (m x n)
        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)

        Return: vector dJ/dw (n x 1) and number dJ/db (1 x 1)
        """
        m = len(y_true)
        dJ_dw = (1/m)*(X.T@(y_pred - y_true))
        dJ_db = (1/m)*(np.sum(y_pred - y_true))
        return (dJ_dw, dJ_db)

    def update_parameters(self, X, y_true, y_pred, w, b, alpha):
        """
        This function updates parameters w and b according to their derivations.
        It should compute the cost function derivations with respect to w and b first,
            then take a step for each parameters in w and b.

        X: the data set matrix (m x n)
        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)
        w: the weights vector (n x 1)
        b: the bias (1 x 1)
        alpha: the learning rate

        Return: the updated parameters w and b
        """
        [dJ_dw, dJ_db] = self.gradient(X , y_true , y_pred) 
        w = w - alpha * dJ_dw 
        b = b - alpha * dJ_db
        return (w ,b)
    
    def create_mini_batches(self, X, y, batch_size):
        mini_batches = []
        data = np.hstack((X, y.reshape(y.shape[0], 1)))
        np.random.shuffle(data)
        n_minibatches = data.shape[0]//batch_size
        i = 0

        for i in range(n_minibatches + 1):
            mini_batch = data[i * batch_size:(i + 1)*batch_size, :]
            X_mini = mini_batch[:, :-1]
            Y_mini = mini_batch[:, -1].reshape((-1, 1))
            mini_batches.append((X_mini, Y_mini))
        if data.shape[0] % batch_size != 0:
            mini_batch = data[i * batch_size:data.shape[0]]
            X_mini = mini_batch[:, :-1]
            Y_mini = mini_batch[:, -1].reshape((-1, 1))
            mini_batches.append((X_mini, Y_mini))
        return mini_batches

    def mini_batch_gradient_descent(self, X, y, alpha = 0.0001, epoch = 100, batch_size = 32):
        """
        The Mini batch gradient descent algorithm:
            1. initialize parameters w and b
            for i in epoch:
                2. compute the hypothesis h(X, w, b)
                3. update the parameters
                4. compute the cost (optional)

        X: the data set matrix (m x (n+1))
        y: the vector of true labels of data items (m x 1)
        alpha: the learning rate
        epoch: number of steps mini-batch GD should take to converge
        batch_size: the number of examples you look at before making a weight update. 

        Returns: the best parameters w and b gradient descent found at last
        """
        speedLog = time()
        w = np.zeros((len(X.T),1))
        b = 0
        costPerEpoch = []
        for i in range(epoch):
            mini_batches = self.create_mini_batches(X, y, batch_size)
            for mini_batch in mini_batches:
                X_mini, y_mini = mini_batch
                hyp = self.h(X_mini, w, b)
                w, b = self.update_parameters(X_mini, y_mini, hyp, w, b, alpha)
                if len(y_mini) == 0:
                    continue
                costPerEpoch.append(self.MSE(y_mini, hyp))
        speedLog = time() - speedLog
        return w, b, costPerEpoch, speedLog
```

```python
class GD_Random_Params:
    def h(self, X, w, b):
        """
        The hypothesis function

        X: the data set matrix (m x n)
        w: the weights vector (n x 1)
        b: the bias (1 x 1)

        Return: a vector stacking predictions for all data items (m x 1)
        """
        return X@w + b

    def MSE(self, y_true, y_pred):
        """
        Mean squared error cost function

        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)

        Return: a single number representing the cost
        """
        return (1/(2*len(y_true)))*np.sum((y_pred-y_true)**2)
    
    def gradient(self, X, y_true, y_pred):
        """
        The gradient of cost function

        X: the data set matrix (m x n)
        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)

        Return: vector dJ/dw (n x 1) and number dJ/db (1 x 1)
        """
        m = len(y_true)
        dJ_dw = (1/m)*(X.T@(y_pred - y_true))
        dJ_db = (1/m)*(np.sum(y_pred - y_true))
        return (dJ_dw, dJ_db)
    
    def update_parameters(self, X, y_true, y_pred, w, b, alpha):
        """
        This function updates parameters w and b according to their derivations.
        It should compute the cost function derivations with respect to w and b first,
            then take a step for each parameters in w and b.

        X: the data set matrix (m x n)
        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)
        w: the weights vector (n x 1)
        b: the bias (1 x 1)
        alpha: the learning rate

        Return: the updated parameters w and b
        """
        [dJ_dw, dJ_db] = self.gradient(X , y_true , y_pred) 
        w = w - alpha * dJ_dw 
        b = b - alpha * dJ_db
        return (w ,b)
    
    def gradient_descent(self, X, y, alpha = 0.0001, epoch = 50):
        """
        The gradient descent algorithm:
            1. initialize parameters w and b
            for i in epoch:
            2. compute the hypothesis h(X, w, b)
            3. update the parameters using function update_parameters(X, y_true, y_pred, w, b, alpha)
            4. compute the cost and see the cost is decreasing in each step (optional)

        X: the data set matrix (m x n)
        y: the vector of true labels of data items (m x 1)
        alpha: the learning rate
        epoch: number of steps gradient descent should take to converge

        Return: the best parameters w and b gradient descent found at last
        """
        speedLog = time()
        w = np.random.randn(len(X.T),1)
        b = np.random.randn()
        costPerEpoch = np.zeros((epoch))
        for i in range(epoch):
            hyp = self.h(X, w, b)
            w, b = self.update_parameters(X, y, hyp, w, b, alpha)
            costPerEpoch[i] = self.MSE(y, hyp)
        speedLog = time() - speedLog
        return w, b, costPerEpoch, speedLog
```

```python
class SGD_MAE:
    def h(self, X, w, b):
        """
        The hypothesis function

        X: the data set matrix (m x n)
        w: the weights vector (n x 1)
        b: the bias (1 x 1)

        Return: a vector stacking predictions for all data items (m x 1)
        """
        return X@w + b

    def MAE(self, y_true, y_pred):
        """
        Mean absolute error cost function

        y_true: the vector of true labels of data items (m x 1)
        y_pred: the vector of predictions of data items (m x 1)

        Return: a single number representing the cost
        """
        return (1/len(y_true))*np.sum(np.absolute(y_pred-y_true))

    def stochastic_gradient_descent(self, X, y, alpha = 0.0001, epoch = 50, batch_size = 1):
        """
        The Stochastic gradient descent algorithm:
            1. initialize parameters w and b
            for i in epoch:
            2. compute the hypothesis h(X, w, b)
            3. update the parameters
            4. compute the cost (optional)

        X: the data set matrix (m x n)
        y: the vector of true labels of data items (m x 1)
        alpha: the learning rate
        epoch: number of steps SGD should take to converge
        batch_size: the number of examples you look at before making a weight update. 

        Return: the best parameters w and b gradient descent found at last
        """    
        speedLog = time()
        w = np.zeros((len(X.T),1))
        b = 0
        costPerEpoch = np.zeros((epoch))
        for i in range(epoch):

            indexes = np.random.randint(0, len(X), batch_size) # random sample

            Xs = X[indexes]
            ys = y[indexes]
            N = len(Xs)
            
            hyp = self.h(Xs, w, b) - ys

            # Updating parameters w and b            
            w = w - alpha * (1/N)*Xs.T@np.sign(hyp)
            b = b - alpha * (1/N)*np.sign(hyp)
            
            costPerEpoch[i] = self.MAE(y, hyp)

        speedLog = time() - speedLog
        return w, b, costPerEpoch, speedLog
```

```python
# run stochastic gradient descent algorithm
w, b, SGD_MAEcostPerEpoch, SGD_MAEspeedLog = SGD_MAE().stochastic_gradient_descent(X, y, 0.00001, 100)
```

```python
# plot SGD results
plt.figure(figsize=(18, 6))

plt.subplot(121)
plt.plot(SGD_MAEcostPerEpoch, 'r-')
plt.xlabel('#Epoch')
plt.ylabel('Cost')
plt.title('SGD with MAE')

plt.subplot(122)
plt.plot(SGDcostPerEpoch, 'r-')
plt.xlabel('#Epoch')
plt.ylabel('Cost')
plt.title('SGD with MSE')

plt.show()
```

```python
def feature_scaling(X):
    """
    Feature Scaling function to normalized data

    X: the data set matrix (m x n)

    Return: the feature scaled data set matrix (m x n)
    """
    return (X - np.mean(X, axis=0))/np.std(X, axis=0)
```

```python
# run stochastic gradient descent algorithm on data WITHOUT feature scaling
w1, b1, SGDcostPerEpoch1, SGDspeedLog1 = SGD().stochastic_gradient_descent(X, y, 0.00001, 100)

# normalize data set matrix
X_normalized = feature_scaling(X)

# run stochastic gradient descent algorithm on data WITH feature scaling
w2, b2, SGDcostPerEpoch2, SGDspeedLog2 = SGD().stochastic_gradient_descent(X_normalized, y, 0.00001, 100)
```
